```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('./data/processed/Somalia_Government_Finances_2013_2024.csv')
```

```python
data.head()
```

```python
data.columns
```

```python
# Define the columns to keep and their new simple, clear names
columns_to_keep = {
    'year': 'year',  # Year of the record
    'month': 'month',  # Month of the record
    'period': 'date',  # Full date of the transaction
    'main': 'category',  # Revenue or Expenditure
    'costcentermaindesc': 'cost_center',  # Description of the cost center
    'name5desc': 'revenue_or_expense_type',  # Type of revenue or expenditure
    'obj4desc': 'tax_or_funding_source',  # Specific source of tax or funds
    'obj6desc': 'payer_or_recipient',  # Who is paying or receiving the money
    'gfs2desc': 'gfs_level_2',  # GFS classification level 2
    'gfs3desc': 'gfs_level_3',  # GFS classification level 3
    'gfs4desc': 'gfs_level_4',  # GFS classification level 4
    'originalbudget': 'budgeted_amount',  # Initial planned budget
    'totalbudget': 'revised_budget',  # Updated total budget
    'actual': 'actual_spending_or_revenue',  # Real amount received or spent
}


# Select and rename columns
data = data[list(columns_to_keep.keys())].rename(columns=columns_to_keep)

print("✅ Columns renamed to simple and clear names successfully!")
```

```python
# Check duplicate rows considering only non-numeric columns
duplicate_check = data.duplicated(subset=['year', 'month', 'date', 'category', 'cost_center', 
                                          'revenue_or_expense_type', 'tax_or_funding_source', 
                                          'payer_or_recipient', 'gfs_level_2', 'gfs_level_3', 
                                          'gfs_level_4'], keep=False)

# Display suspected duplicates
data[duplicate_check]
```

```python
# Convert '-' to NaN in numerical columns
num_cols = ['budgeted_amount', 'revised_budget', 'actual_spending_or_revenue']
data[num_cols] = data[num_cols].replace('-', np.nan)

# Convert columns to float
data[num_cols] = data[num_cols].astype(float)
```

```python
from sklearn.impute import KNNImputer

# Select numerical columns
num_cols = ['budgeted_amount', 'revised_budget', 'actual_spending_or_revenue']

# Initialize and apply KNN imputer
knn_imputer = KNNImputer(n_neighbors=5)  # Using 5 nearest neighbors
data[num_cols] = knn_imputer.fit_transform(data[num_cols])

print("✅ Numerical missing values imputed using KNN!")
```

```python
# Fill missing categorical values with mode (most frequent value)
cat_cols = ['cost_center', 'category','revenue_or_expense_type', 'tax_or_funding_source',
            'payer_or_recipient', 'gfs_level_2', 'gfs_level_3', 'gfs_level_4']

for col in cat_cols:
    data[col].fillna(data[col].mode()[0], inplace=True)

print("✅ Categorical missing values filled with mode!")
```
